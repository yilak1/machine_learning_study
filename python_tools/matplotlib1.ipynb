# Matplotlib基础

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0, 10, 100)
```

```python
x
```

```python
y = np.sin(x)
```

```python
plt.plot(x, y)
```

```python
plt.show()
```

```python
cosy = np.cos(x)
```

```python
cosy.shape
```

```python
siny = y.copy()
```

```python
plt.plot(x, siny)
plt.plot(x, cosy)
```

```python
plt.plot(x, siny)
plt.plot(x, cosy, color="red")
```

```python
plt.plot(x, siny)
plt.plot(x, cosy, color="red", linestyle=':')
```

```python
plt.plot(x, siny)
plt.plot(x, cosy, color="red", linestyle=':')
plt.xlim(-5, 15)
plt.ylim(0, 1.5)
```

```python
plt.plot(x, siny)
plt.plot(x, cosy, color="red", linestyle=':')
plt.axis([-1, 11, -2, 2])
```

```python
plt.plot(x, siny)
plt.plot(x, cosy, color="red", linestyle=':')
plt.xlabel("x-axis")
plt.ylabel("y-value")
```

```python
plt.plot(x, siny, label="sin(x)")
plt.plot(x, cosy, color="red", linestyle=':', label="cos(x)")
plt.xlabel("x-axis")
plt.ylabel("y-value")
plt.legend()
plt.title("Welcome")
```
